# First Chat Completion with Telnyx Inference

In this notebook, you’ll send your first chat completion request to a Telnyx-hosted model using the OpenAI Python SDK.

Telnyx Inference is OpenAI-compatible, so if you already use the OpenAI SDK, you can point it at Telnyx by changing the `base_url` and using your Telnyx API key.

## Add Telnyx API key

```python
from getpass import getpass

TELNYX_API_KEY = getpass("Enter your Telnyx API key: ")
```

## Create Client

```python
from openai import OpenAI

client = OpenAI(
    api_key=TELNYX_API_KEY,
    base_url="https://api.telnyx.com/v2/ai/openai"
)
```

## First Chat Completion

```python
response = client.chat.completions.create(
    model="moonshotai/Kimi-K2.6",
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant."
        },
        {
            "role": "user",
            "content": "Explain Telnyx Inference in one short paragraph."
        }
    ],
    temperature=0.2,
)

print(response.choices[0].message.content)
```

## Try Your Own Prompt

Change the prompt below and run the cell again.


```python
user_prompt = "Give me three example use cases for low-latency LLM inference."

response = client.chat.completions.create(
    model="moonshotai/Kimi-K2.6",
    messages=[
        {"role": "system", "content": "You are a concise technical assistant."},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.2,
)

print(response.choices[0].message.content)
```
